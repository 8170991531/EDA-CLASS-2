```python
#category = [ 'Type', 'Content Rating'] wise count plot
# categorical columns
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold',
alpha=0.8, y=1.)
category = [ 'Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=dk[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
# this is multivariate analysis.
# Q4: what are the top 5 most installed Apps in each popular Category.
dka=dk.groupby(['Category','App'])['Installs'].sum().reset_index()
```

```python
dka
```

```python
dka=dka.sort_values('Installs',ascending=False)
```

```python
# for different different category.
dka[dka.Category=='GAME']
```

```python
# for five category in GAME.
dka[dka.Category=='GAME'][:5]
```

```python
dka[dka.Category=='PRODUCTIVITY'][:5]
```

```python
dka[dka.Category=='FAMILY'][:5]
```

```python
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")
plt.figure(figsize=(40,30))
for i,app in enumerate(apps):
    dk2 = dka[dka.Category == app]
    dk3 = dk2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= dk3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```
